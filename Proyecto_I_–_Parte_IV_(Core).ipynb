```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el dataset
df = pd.read_csv('retail_sales_dataset.csv')

print("\nInformación del DataFrame:")
df.info()

# Mostrar las primeras 5 filas del DataFrame para verificar que se cargó correctamente
print("Primeras 5 filas del DataFrame:")
print(df.head())

# Calcular estadísticas descriptivas para las columnas numéricas
print("\nEstadísticas descriptivas básicas:")
print(df.describe())
```

```python
# Convertir la columna 'Date' a formato datetime
df['Date'] = pd.to_datetime(df['Date'])
print("Datos cargados y la columna 'Date' convertida a tipo datetime.")
print("\nInformación del DataFrame:")
df.info()
```

```python
import matplotlib.pyplot as plt

# Histograma de la columna 'Total Amount'
plt.figure(figsize=(10, 6))
plt.hist(df['Total Amount'], bins=30, edgecolor='black', alpha=0.7)
plt.title('Distribución de los Montos Totales de Venta')
plt.xlabel('Monto Total')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()

# Boxplot de la columna 'Total Amount'
plt.figure(figsize=(8, 6))
plt.boxplot(df['Total Amount'])
plt.title('Boxplot de los Montos Totales de Venta')
plt.ylabel('Monto Total')
plt.grid(True)
plt.show()
```

```python
# Agrupar por fecha y sumar los montos totales
ventas_por_dia = df.groupby('Date')['Total Amount'].sum()

# Crear un gráfico de líneas
plt.figure(figsize=(15, 7))
plt.plot(ventas_por_dia.index, ventas_por_dia.values, marker='o', linestyle='-')
plt.title('Tendencia de Ventas a lo Largo del Tiempo')
plt.xlabel('Fecha')
plt.ylabel('Monto Total de Ventas')
plt.grid(True)
plt.show()
```

```python
# Gráfico de dispersión entre 'Age' y 'Total Amount'
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['Total Amount'], alpha=0.5)
plt.title('Relación entre Edad y Monto Total de Venta')
plt.xlabel('Edad del Cliente')
plt.ylabel('Monto Total de Venta')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns # seaborn es una biblioteca que facilita la creación de gráficos estadísticos atractivos

# Crear una figura con dos subplots
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Histograma en el primer subplot
sns.histplot(df['Total Amount'], bins=30, kde=True, ax=axes[0])
axes[0].set_title('Distribución de Montos de Venta (Histograma)')
axes[0].set_xlabel('Monto Total')
axes[0].set_ylabel('Frecuencia')

# Boxplot en el segundo subplot
sns.boxplot(x=df['Total Amount'], ax=axes[1])
axes[1].set_title('Distribución de Montos de Venta (Boxplot)')
axes[1].set_xlabel('Monto Total')

plt.tight_layout() # Ajusta automáticamente los parámetros de la subtrama
plt.show()
```
